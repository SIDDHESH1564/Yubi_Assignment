```python
# Start completely fresh after restarting the runtime
!pip install numpy==1.19.5
!pip install librosa==0.8.0
!pip install opencv-python==4.5.4.60
!pip install gtts
!apt-get update -qq
!apt-get install -y -qq ffmpeg

# Clone Wav2Lip
!git clone https://github.com/Rudrabha/Wav2Lip
%cd Wav2Lip

# Fix the np.complex issue in audio.py
!sed -i 's/np.complex/np.complex128/g' audio.py

# Set up model directories
!mkdir -p checkpoints
!wget -nc https://github.com/Rudrabha/Wav2Lip/releases/download/v1.0/wav2lip_gan.pth -O checkpoints/wav2lip_gan.pth

!mkdir -p face_detection/detection/sfd/
!wget -nc https://www.adrianbulat.com/downloads/python-fan/s3fd-619a316812.pth -O face_detection/detection/sfd/s3fd.pth

# Generate audio
from gtts import gTTS

script = """Namaste Mathangi! My name is Anika, and I'm here to guide you through managing your credit
card dues. Mathangi, as of today, your credit card bill shows an amount due of INR 5,053 which
needs to be paid by 31st December 2024
Missing this payment could lead to two significant consequences:
First, a late fee will be added to your outstanding balance.
Second, your credit score will be negatively impacted, which may affect your future borrowing
ability.
Make your payment by clicking the link here... Pay through UPI or bank transfer. Thank you!"""

tts = gTTS(text=script, lang='en', tld='co.in')
tts.save('audio.mp3')
!ffmpeg -i audio.mp3 -acodec pcm_s16le -ar 16000 audio.wav -y

# Upload image
from google.colab import files
print("Please upload the image for lip syncing:")
uploaded = files.upload()
image_file = list(uploaded.keys())[0]

# Run inference with minimal parameters
!python inference.py --checkpoint_path checkpoints/wav2lip_gan.pth --face "{image_file}" --audio audio.wav --outfile result_video.mp4 --static

# Display result
from IPython.display import HTML
from base64 import b64encode
import os

if os.path.exists('result_video.mp4'):
    with open('result_video.mp4', 'rb') as f:
        video = f.read()
    video_url = f"data:video/mp4;base64,{b64encode(video).decode()}"
    display(HTML(f"""<video width=400 controls><source src="{video_url}"></video>"""))
    files.download('result_video.mp4')
    print("Success! Video generated and ready for download.")
else:
    print("Video generation failed. See error messages above.")
```
